````python
tg['data'] = pd.to_datetime(tg['data'])
````

````python
merged_df = df.merge(
    tg[['data', 'curs']].rename(columns={'data': 'order_dt'}),
    on='order_dt',
    how='left'
)
merged_df.head()
````

````python
def convert_to_rub(row):
    if row['currency_code'] == 'rub':
        return row['revenue']
    elif row['currency_code'] == 'kzt':
        return (row['revenue'] / 100) * row['curs']
    else:
        return None 

merged_df['revenue_rub'] = merged_df.apply(convert_to_rub, axis=1)
````

    Когда нам нужно провести агрегацию через несколько уровней (к примеру, наш случай, сначала до пользователей, затем общую), то на этапе второй агрегации у нас появляется два пути - посчитать общее среднее или среднее от средних. **Среднее от средних** подразумевает, что мы берем уже готовые средние показатели первого этпа агрегации (по каждому пользователю) и усредняем, тем самым как бы оценивая стоимость заказа, которую готов заплатить **среднестатистический пользователь**. При интерпретации речь, разумеется, тоже должна идти о пользователях. В свою очередь **общее среднее** подразумевает - раз мы считаем среднюю стоимость заказа, то и работать мы должны с заказами - в числителе берем суммарную стоимость заказов, в знаменателе берем суммарное количество заказов. В этом случае мы уже оцениваем величину стоимости **среднестатистического заказа**. Результаты по этим двум подходам будут отличаться. Причем, чем разнороднее наши пользователи по объему заказов и затрат, тем существеннее будут разниться результаты этих двух подходов, поскольку при расчете среднего от средних мы уравниваем вклад каждого пользователя, влияние экстремальных пользователей с большим количеством заказов на показатель снижается (в общем среднем они продавливают показатель своей кучей заказов), а влияние пользователей с небольшим количеством заказов, напротив, нарастает (в общем среднем они теряются среди остальных, задавливаются).

Изучите, какие характеристики первого заказа и профиля пользователя могут быть связаны с числом покупок. Для этого используйте универсальный коэффициент корреляции `phi_k`, который позволяет анализировать как числовые, так и категориальные признаки.

---

**Задача 4.4.1:** Проведите корреляционный анализ:
- Рассчитайте коэффициент корреляции `phi_k` между признаками профиля пользователя и числом заказов (`total_orders`). При необходимости используйте параметр `interval_cols` для определения интервальных данных.
- Проанализируйте полученные результаты. Если полученные значения будут близки к нулю, проверьте разброс данных в `total_orders`. Такое возможно, когда в данных преобладает одно значение: в таком случае корреляционный анализ может показать отсутствие связей. Чтобы этого избежать, выделите сегменты пользователей по полю `total_orders`, а затем повторите корреляционный анализ. Выделите такие сегменты:
    - 1 заказ;
    - от 2 до 4 заказов;
    - от 5 и выше.
- Визуализируйте результат корреляции с помощью тепловой карты.
- Ответьте на вопрос: какие признаки наиболее связаны с количеством заказов?

````python
interval_cols = [
    'avg_revenue_per_order',
    'avg_tickets_per_order', 
    'avg_time_between',
    'avg_revenue_rub']

categorical_cols = [
    'device_type_canonical',
    'region_name', 
    'service_name',
    'event_type_main',
    'weekday']

target_col = 'total_orders'

cols_for_analysis = interval_cols + categorical_cols + [target_col]
df_analysis = df_filtered_95_copy[cols_for_analysis].copy()

correlation_matrix = df_analysis.phik_matrix(interval_cols=interval_cols)
total_orders_corr = correlation_matrix[target_col].sort_values(ascending=False)

print("Корреляция признаков с числом заказов:")
print(total_orders_corr)
````

````python
plt.figure(figsize=(8, 10))
sns.heatmap(
    total_orders_corr.to_frame(name='Phi-K'), 
    annot=True, 
    fmt=".3f", 
    cmap='coolwarm', 
    vmin=0, 
    vmax=1,
    cbar_kws={'label': 'Phi-K коэффициент'}
)
plt.title('Phi-K корреляция признаков с числом заказов', fontsize=14)
plt.ylabel('Признак')
plt.xlabel('Коэффициент корреляции')
plt.tight_layout()
plt.show()
````

**Вывод**: На графике видно что **наибольшая связь с количеством заказов у среднего времени между заказами: чем меньше среднее время между заказами, тем больше всего заказов делает пользователь**. Пользователи, которые берут больше билетов за раз, в среднем делают и больше заказов (умеренная связь). Размер среднего чека коррелирует с частотой заказов (но не обязательно линейно - умеренная связь). Нет связи между днем недели, регионом с числом заказов. Почти нет связи с билетным оператором, типом мероприятия и устройством, с которого сделан заказ.

````python
plt.figure(figsize=(10, 6))
sns.countplot(x='total_orders', data=df_analysis, order=df_analysis['total_orders'].value_counts().index)
plt.title('Распределение числа заказов (total_orders)')
plt.xlabel('Количество заказов')
plt.ylabel('Число пользователей')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()

print("Доля пользователей с 1 заказом:", df_analysis['total_orders'].eq(1).mean())
print("Доля пользователей с 2-4 заказами:", ((df_analysis['total_orders'] >= 2) & (df_analysis['total_orders'] <= 4)).mean())
print("Доля пользователей с 5+ заказами:", (df_analysis['total_orders'] >= 5).mean())
````

````python
df_analysis['is_loyal'] = (df_analysis['total_orders'] >= 5).astype(int)

interval_cols = [
    'avg_revenue_per_order',
    'avg_tickets_per_order', 
    'avg_time_between',
    'avg_revenue_rub']

categorical_cols = [
    'device_type_canonical',
    'region_name', 
    'service_name',
    'event_type_main',
    'weekday']

cols_for_phik = interval_cols + categorical_cols + ['is_loyal']
corr_matrix = df_analysis[cols_for_phik].phik_matrix(interval_cols=interval_cols)
loyalty_corr = corr_matrix['is_loyal'].sort_values(ascending=False)

print("Корреляция признаков с меткой 'Лояльный пользователь (5+ заказов)':")
print(loyalty_corr)

plt.figure(figsize=(8, 10))
sns.heatmap(
    loyalty_corr.to_frame(name='Phi-K с is_loyal'), 
    annot=True, 
    fmt=".3f", 
    cmap='coolwarm', 
    vmin=0, 
    vmax=1,
    cbar_kws={'label': 'Сила связи с лояльностью'}
)
plt.title('Какие признаки лучше всего предсказывают попадание в сегмент "5+ заказов"', fontsize=14)
plt.ylabel('Признак')
plt.xlabel('Коэффициент Phi-K')
plt.tight_layout()
plt.show()
````

````python
segments = {
    '1 заказ': df_analysis[df_analysis['total_orders'] == 1],
    '2–4 заказа': df_analysis[(df_analysis['total_orders'] >= 2) & (df_analysis['total_orders'] <= 4)],
    '5+ заказов': df_analysis[df_analysis['total_orders'] >= 5]}

cols = interval_cols + categorical_cols

for seg_name, seg_df in segments.items():
    if len(seg_df) < 50:
        print(f"Пропускаем сегмент '{seg_name}': слишком мало данных ({len(seg_df)} строк)")
        continue
        
    print(f"\n--- Матрица корреляций внутри сегмента: {seg_name} ---")
    
    seg_corr = seg_df[cols].phik_matrix(interval_cols=interval_cols)
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(seg_corr, annot=True, fmt=".2f", cmap='coolwarm', vmin=0, vmax=1)
    plt.title(f'Phi-K корреляции признаков внутри сегмента "{seg_name}"')
    plt.tight_layout()
    plt.show()
````

**Вывод:** Для анализа связи признаков с уровнем лояльности (количеством заказов) была использована сегментация пользователей. Поскольку прямая корреляция внутри сегментов невозможна из-за низкой дисперсии целевой переменной, был использован бинарный индикатор лояльности (is_loyal_5plus = 1 при 5+ заказах).

Результаты расчёта коэффициента Phi-K показали:
1. `avg_time_between` (коэф. ~0.5): Наиболее значимый фактор. Пользователи, делающие заказы с минимальными интервалами, с высокой вероятностью переходят в сегмент 5+ заказов.
2. `service_name` (коэф. ~0.25): Тип сервиса является вторым по значимости фактором. Некоторые сервисы демонстрируют значительно более высокую склонность пользователей к формированию долгосрочной лояльности.
3. `avg_tickets_per_order` (коэф. ~0.2): Объём покупки (количество билетов в чеке) также коррелирует с лояльностью, указывая на то, что активные пользователи склонны покупать больше билетов за одну транзакцию.

**Вывод:** 

Анализ матрицы Phi‑K **внутри сегмента пользователей с 1 заказом показал, что поведение новичков неоднородно и сильно зависит от географического фактора**. Наиболее сильная связь наблюдается между `region_name` и `avg_time_between` (коэф. 0.73), а также между `region_name` и `service_name` (коэф. 0.67). Это означает, что стратегия удержания должна быть региональной: в одних регионах пользователи после первого заказа быстро теряют интерес, а в других — остаются активными дольше.Тип устройства и день недели не оказывают значимого влияния на поведение новичков, поэтому не должны быть основными критериями для таргетирования.

**Вывод:**

Наиболее сильная связь наблюдается между `region_name` и `service_name` (коэф. 0.68), что указывает на жёсткую региональную специфику потребления. Тип устройства и день недели не оказывают влияния на поведение пользователей в этом сегменте.
Таким образом, для увеличения количества заказов у пользователей с 2–4 покупками необходимо сфокусироваться на региональной адаптации продуктовых механик и сокращении интервала между заказами через персонализированные триггерные коммуникации, учитывающие специфику выбранного сервиса.

1. Создайте файл `.gitignore`. Добавьте в него все временные и чувствительные файлы, которые не должны попасть в репозиторий.
2. Сформируйте файл `requirements.txt`. Зафиксируйте все библиотеки, которые вы использовали в проекте.
3. Вынести все чувствительные данные (параметры подключения к базе) в `.env`файл.
4. Проверьте, что проект запускается и воспроизводим.
5. Загрузите проект в публичный репозиторий — например, на GitHub. Убедитесь, что все нужные файлы находятся в репозитории, исключая те, что в `.gitignore`. Ссылка на репозиторий понадобится для отправки проекта на проверку. Вставьте её в шаблон проекта в тетрадке Jupyter Notebook перед отправкой проекта на ревью.

<b>Рекомендации💡:</b> Сейчас работу с Git сделали необязательной, но напомню про довольно критичные моменты.

* <span style="color:Red;">В репозитории не должно быть опубликовано данных и ключей для доступа к ним (как в `.env`, так и в самой тетрадке, **в том числе в тексте задания и комментариях**), это считается засекреченной информацией. Можно оставить лишь `env_exaple.txt`, чтобы читатели, желающие запустить код, вставили туда свои собственные ключи. Сам `.env` необходимо отправить в игнорирование. (Обрати только внимание, что если ключи уже были случайно опубликованы, то после удаления из репозитория они все равно останутся в истории коммитов, поэтому ее тоже нужно почистить, хотя это и не очень простой процесс.) </span>
    
    
* Публиковать лучше тетрадку с выполненными ячейками. Просто код довольно бесполезен в контексте анализа, поскольку читатель не сможет ее запустить самостоятельно, так как у него нет доступа к данным.


* Стоит заполнить `readme` полезной информацией о проекте, зафиксировать цели, указать ключевые моменты и полученные выводы.
